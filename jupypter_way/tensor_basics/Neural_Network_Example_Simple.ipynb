<h1>What is Neural Network</h1>

<li>A neural network contains input and outputs</li>
<li><font color="blue">Error is added to the output</font></li>
<li><font color="red">Computational graph is run multiple times</font> using <font color="blue">GradientDescentOptimizer</font></li>
<li>The <font color="green">formula where loss is minimal</font> is considered to be the best fit</li>

## A simple neural network
## Purpose: Ccompute a linear regression between 2 variables, y and x

<li>The function it tries to compute is the best $w_1$ and $w_2$ it can find for the function <font color="blue">$y = w_2 x + w_1$</font> for the data </li>
<li>The data we're going to give it is <b>toy data</b>, <font color="red">linear perturbed with random noise</font></li>

Here is the TensorFlow code for this simple neural network and the results of running this code:

```python
#@test {"output": "ignore"}
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# inline means print in the current backend, viz, the jupyter notebook in this example
%matplotlib inline

# Set up the data with a noisy linear relationship between X and Y.
num_examples = 50
# np.linspace generates linearly spaced vectors
#   cens.ioc.ee/local/man/matlab/techdoc/ref/linspace.html
# Here 50 points are generated between -2 and 4 and 50 points are generated between -6 and 6
X = np.array([np.linspace(-2, 4, num_examples), np.linspace(-6, 6, num_examples)])
X += np.random.randn(2, num_examples)
x, y = X
bias_with_x = np.array([(1., a) for a in x]).astype(np.float32)

losses = []
training_steps = 50
learning_rate = 0.002

with tf.Session() as sess:
    # Set up all the tensors, variables, and operations.
    input = tf.constant(bias_with_x)
    # transpose swaps the rows and columns of the array
    target = tf.constant(np.transpose([y]).astype(np.float32))
    weights = tf.Variable(tf.random_normal([2, 1], 0, 0.1))

    tf.global_variables_initializer().run()

    yhat = tf.matmul(input, weights)
    yerror = tf.subtract(yhat, target)
    loss = tf.nn.l2_loss(yerror)
  
    update_weights = tf.train.GradientDescentOptimizer(learning_rate).minimize(loss)
  
    for _ in range(training_steps):
        # Repeatedly run the operations, updating the TensorFlow variable.
        update_weights.run()
        losses.append(loss.eval())

    # Training is done, get the final values for the graphs
    betas = weights.eval()
    yhat = yhat.eval()

# Show the fit and the loss over time.
fig, (ax1, ax2) = plt.subplots(1, 2)
plt.subplots_adjust(wspace=.3)
fig.set_size_inches(10, 4)
ax1.scatter(x, y, alpha=.7)
ax1.scatter(x, np.transpose(yhat)[0], c="g", alpha=.6)
line_x_range = (-4, 6)
ax1.plot(line_x_range, [betas[0] + a * betas[1] for a in line_x_range], "g", alpha=0.6)
ax2.plot(range(0, training_steps), losses)
ax2.set_ylabel("Loss")
ax2.set_xlabel("Training steps")
plt.show()
```

# Code in Detail

This is a toy data set here. We have 50 (x,y) data points. At first, the data is perfectly linear.

```python
#@test {"output": "ignore"}
num_examples = 50
# np.linspace generates linearly spaced vectors
#   cens.ioc.ee/local/man/matlab/techdoc/ref/linspace.html
# Here 50 points are generated between -2 and 4 and 50 points are generated between -6 and 6
X = np.array([np.linspace(-2, 4, num_examples), np.linspace(-6, 6, num_examples)])
plt.figure(figsize=(4,4))
# scatter is similar to plot function
# See scatter_example
plt.scatter(X[0], X[1])
plt.show()
```

<h1>We add noise now</h1>

```python
#@test {"output": "ignore"}
X += np.random.randn(2, num_examples)
plt.figure(figsize=(4,4))
plt.scatter(X[0], X[1])
plt.show()
```

## What we want to do

What we're trying to do is calculate the green line below:

```python
#@test {"output": "ignore"}
weights = np.polyfit(X[0], X[1], 1)
plt.figure(figsize=(4,4))
plt.scatter(X[0], X[1])
line_x_range = (-3, 5)
plt.plot(line_x_range, [weights[1] + a * weights[0] for a in line_x_range], "g", alpha=0.8)
plt.show()
```

That's equivalent to the function $\hat{y} = w_2 x + w_1$. What we're trying to do is find the "best" weights $w_1$ and $w_2$. That will give us that green regression line above.

<font color="blue"><h2>What are the best weights aka L2 Loss</h2></font>
<li>Weights that <font color="green">minimize the difference between our estimate $\hat{y}$ and the actual y</font></li?
<li>Specifically, we want to <font color="red">minimize the sum of the squared errors</font></li>
<li>So minimize $\sum{(\hat{y} - y)^2}$, which is known as the *L2 loss*</li> 
<li>So, the best weights are the weights that minimize the L2 loss

<h2><font color="blue">Gradient descent</font></h2>

What gradient descent does is <font color="red">start with random weights</font> for $\hat{y} = w_2 x + w_1$ and gradually <font color="red"> moves those weights toward better values</font>

It does that by following the downward slope of the error curves. Imagine that the possible errors we could get with different weights as a landscape. From whatever weights we have, <font color="blue">moving in some directions will increase the error, like going uphill, and some directions will decrease the error</font>, like going downhill. We want to roll downhill, always moving the weights toward lower error.

How does gradient descent know which way is downhill? It follows the partial derivatives of the <font color="red">L2 loss</font>. The partial derivative is like a velocity, saying which way the error will change if we change the weight. <font color="green"><b>We want to move in the direction of lower error</font></b>. The partial derivative points the way.

So, what gradient descent does is start with random weights and gradually walk those weights toward lower error, using the partial derivatives to know which direction to go

<h1><font color="red">The code again</font></h1>

```python
#@test {"output": "ignore"}
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Set up the data with a noisy linear relationship between X and Y.
num_examples = 50
X = np.array([np.linspace(-2, 4, num_examples), np.linspace(-6, 6, num_examples)])
# Add random noise (gaussian, mean 0, stdev 1)
X += np.random.randn(2, num_examples)
# Split into x and y
x, y = X
# Add the bias node which always has a value of 1
bias_with_x = np.array([(1., a) for a in x]).astype(np.float32)

# Keep track of the loss at each iteration so we can chart it later
losses = []
# How many iterations to run our training
training_steps = 50
# The learning rate. Also known has the step size. This changes how far
# we move down the gradient toward lower error at each step. Too large
# jumps risk inaccuracy, too small slow the learning.
learning_rate = 0.002

# In TensorFlow, we need to run everything in the context of a session.
with tf.Session() as sess:
    # Set up all the tensors.
    # Our input layer is the x value and the bias node.
    input = tf.constant(bias_with_x)
    # Our target is the y values. They need to be massaged to the right shape.
    target = tf.constant(np.transpose([y]).astype(np.float32))
    # Weights are a variable. They change every time through the loop.
    # Weights are initialized to random values (gaussian, mean 0, stdev 0.1)
    weights = tf.Variable(tf.random_normal([2, 1], 0, 0.1))

    # Initialize all the variables defined above.
    tf.global_variables_initializer().run()

    # Set up all operations that will run in the loop.
    # For all x values, generate our estimate on all y given our current
    # weights. So, this is computing y = w2 * x + w1 * bias
    yhat = tf.matmul(input, weights)
    # Compute the error, which is just the difference between our 
    # estimate of y and what y actually is.
    yerror = tf.subtract(yhat, target)
    # We are going to minimize the L2 loss. The L2 loss is the sum of the
    # squared error for all our estimates of y. This penalizes large errors
    # a lot, but small errors only a little.
    loss = tf.nn.l2_loss(yerror)

    # Perform gradient descent. 
    # This essentially just updates weights, like weights -= grads * learning_rate
    # using the partial derivative of the loss with respect to the
    # weights. It's the direction we want to go to move toward lower error.
    update_weights = tf.train.GradientDescentOptimizer(learning_rate).minimize(loss)

    # Now, repeatedly call run, inducing the gradient descent operation
    #   Which repeatedly changes weights by a small amount in the direction to reduce the error (L2 loss)
    for _ in range(training_steps):
        # Repeatedly run the operations, updating the TensorFlow variable.
        sess.run(update_weights)

        # Here, we're keeping a history of the losses to plot later
        # so we can see the change in loss as training progresses.
        losses.append(loss.eval())

    # Training is done, get the final values for the charts
    betas = weights.eval()
    yhat = yhat.eval()

# Show the results.
fig, (ax1, ax2) = plt.subplots(1, 2)
plt.subplots_adjust(wspace=.3)
fig.set_size_inches(10, 4)
ax1.scatter(x, y, alpha=.7)
ax1.scatter(x, np.transpose(yhat)[0], c="g", alpha=.6)
line_x_range = (-4, 6)
ax1.plot(line_x_range, [betas[0] + a * betas[1] for a in line_x_range], "g", alpha=0.6)
ax2.plot(range(0, training_steps), losses)
ax2.set_ylabel("Loss")
ax2.set_xlabel("Training steps")
plt.show()
```

The core piece is the loop, which contains a single `run` call. `run` executes the operations necessary forthe `GradientDescentOptimizer` operation

* The `GradientDescentOptimizer` execution has a <font color="red">side effect of assigning to weights</font>, so the variable weights changes each time in the loop.

<font color="red">After many iterations of the loop, the amount we are changing the weights gets smaller and smaller, and the loss gets smaller and smaller, as we narrow in on near optimal values for the weights</font>

<font color="green">By the end of the loop, we should be near the lowest possible values for the L2 loss, and near the best possible weights we could have</font>

<font color="green">Why this code: To understand better without using above functions</font>

This code looks very similar to the code above, but <font color="red">without using <b>l2_loss</b> or <b>GradientDescentOptimizer</b>

<font color="gray">
<li>loss = 0.5 * tf.reduce_sum(tf.multiply(yerror, yerror))
<br/>
<b>Calculates the L2 loss manually</b>

<li>gradient = tf.reduce_sum(tf.transpose(tf.multiply(input, yerror)), 1, keep_dims=True)

<b><br/>It's the partial derivatives of the L2 loss with respect to $w_1$ and $w_2$, the same thing as whatgradients(loss, weights) does in the earlier code</b>

<li>update_weights = tf.assign_sub(weights, learning_rate * gradient)</li>
<b><br/>Subtracts a constant the gradient after scaling by the learning rate (to avoid jumping too far each time, which risks moving in the wrong direction)</b>
</font>

```python
#@test {"output": "ignore"}

# Use the same input data and parameters as the examples above.
# We're going to build up a list of the errors over time as we train to display later.
losses = []

with tf.Session() as sess:
    # Set up all the tensors.
    # The input is the x values with the bias appended on to each x.
    input = tf.constant(bias_with_x)
    # We're trying to find the best fit for the target y values.
    target = tf.constant(np.transpose([y]).astype(np.float32))
    # Let's set up the weights randomly
    weights = tf.Variable(tf.random_normal([2, 1], 0, 0.1))

    tf.global_variables_initializer().run()

    # learning_rate is the step size, so how much we jump from the current spot
    learning_rate = 0.002

    # The operations in the operation graph.
    # Compute the predicted y values given our current weights
    yhat = tf.matmul(input, weights)
    # How much does this differ from the actual y?
    yerror = tf.subtract(yhat, target)
    # Change the weights by subtracting derivative with respect to that weight
    loss = 0.5 * tf.reduce_sum(tf.multiply(yerror, yerror))
    gradient = tf.reduce_sum(tf.transpose(tf.multiply(input, yerror)), 1, keep_dims=True)
    update_weights = tf.assign_sub(weights, learning_rate * gradient)
    
    # Repeatedly run the operation graph over the training data and weights.
    for _ in range(training_steps):
        sess.run(update_weights)
    
        # Here, we're keeping a history of the losses to plot later
        # so we can see the change in loss as training progresses.
        losses.append(loss.eval())

    # Training is done, compute final values for the graph.
    betas = weights.eval()
    yhat = yhat.eval()

# Show the results.
fig, (ax1, ax2) = plt.subplots(1, 2)
plt.subplots_adjust(wspace=.3)
fig.set_size_inches(10, 4)
ax1.scatter(x, y, alpha=.7)
ax1.scatter(x, np.transpose(yhat)[0], c="g", alpha=.6)
line_x_range = (-4, 6)
ax1.plot(line_x_range, [betas[0] + a * betas[1] for a in line_x_range], "g", alpha=0.6)
ax2.plot(range(0, training_steps), losses)
ax2.set_ylabel("Loss")
ax2.set_xlabel("Training steps")
plt.show()
```
